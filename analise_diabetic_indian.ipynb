```python
import pandas as pd
import matplotlib.pyplot as plt 
import numpy as np
```

```python
dgA = pd.read_csv('archives_data/B. Disease Grading/2. Groundtruths/a. IDRiD_Disease Grading_Training Labels.csv')
```

```python
dgB = pd.read_csv('archives_data/B. Disease Grading/2. Groundtruths/b. IDRiD_Disease Grading_Testing Labels.csv')
```

```python
count_diagnosis_test = dgAB['Retinopathy grade'].value_counts().sort_index()

cores = plt.cm.Reds(np.linspace(0.2, 1, 5)) 
count_diagnosis_test.plot(kind='bar', color=cores)

plt.title('ClassificaÃ§Ã£o das imagens analisadas Indian')
plt.xlabel('ClassificaÃ§Ã£o')
plt.ylabel('Quantidade de imagens')

niveis_dr = ['NÃ£o possui RD', 'Leve', 'Moderada', 'Severa', 'RD Proliferativa']

plt.xticks(range(len(niveis_dr)), niveis_dr, rotation='horizontal')
```

```python
dgAB.rename(columns={'Image name': 'Image No'}, inplace=True)
```

```python
dados_mesclados = OpticData.merge(dtFovea, on='Image No').merge(dgAB, on='Image No')
```

```python
dados_mesclados.rename(columns={'X- Coordinate_x': 'opticX', 'Y - Coordinate_x': 'opticY'}, inplace=True)
dados_mesclados.rename(columns={'X- Coordinate_y': 'foveaX', 'Y - Coordinate_y': 'foveaY'}, inplace=True)
dados_mesclados.rename(columns={'Retinopathy grade': 'grade', 'Risk of macular edema ': 'risk'}, inplace=True)
```

```python
# Definir uma paleta de cores de vermelho claro a vermelho escuro
paleta = plt.cm.Reds(np.linspace(0.3, 1, 5))

# GrÃ¡fico de dispersÃ£o das coordenadas com classificaÃ§Ã£o e risco
plt.figure(figsize=(12, 6))

# Disco Ãptico
plt.subplot(1, 2, 1)
scatter = plt.scatter(dados_mesclados['opticX'], dados_mesclados['opticY'],
                      c=dados_mesclados['grade'], cmap=plt.cm.Reds, alpha=0.6,
                      s=(dados_mesclados['risk'] + 1) * 20, edgecolors='w')
plt.title('Disco Ãptico - ClassificaÃ§Ã£o e Risco de Edema')
plt.xlabel('Coordenada X do Disco Ãptico')
plt.ylabel('Coordenada Y do Disco Ãptico')
cbar = plt.colorbar(scatter, ticks=[0, 1, 2, 3, 4])
cbar.set_label('ClassificaÃ§Ã£o da Retinopatia')
cbar.set_ticks([0, 1, 2, 3, 4])
cbar.set_ticklabels(['0', '1', '2', '3', '4'])
plt.legend(*scatter.legend_elements("sizes", num=3), title='Risco de Edema')

# FÃ³vea
plt.subplot(1, 2, 2)
scatter = plt.scatter(dados_mesclados['foveaX'], dados_mesclados['foveaY'],
                      c=dados_mesclados['grade'], cmap=plt.cm.Reds, alpha=0.6,
                      s=(dados_mesclados['risk'] + 1) * 20, edgecolors='w')
plt.title('FÃ³vea - ClassificaÃ§Ã£o e Risco de Edema')
plt.xlabel('Coordenada X da FÃ³vea')
plt.ylabel('Coordenada Y da FÃ³vea')
cbar = plt.colorbar(scatter, ticks=[0, 1, 2, 3, 4])
cbar.set_label('ClassificaÃ§Ã£o da Retinopatia')
cbar.set_ticks([0, 1, 2, 3, 4])
cbar.set_ticklabels(['0', '1', '2', '3', '4'])
plt.legend(*scatter.legend_elements("sizes", num=3), title='Risco de Edema')

plt.tight_layout()
plt.show()
```

```python
# Criar uma tabela resumo
tabela_resumo = dados_mesclados.groupby(['grade', 'risk']).agg({
    'opticX': 'mean',
    'opticY': 'mean',
    'foveaX': 'mean',
    'foveaY': 'mean'
}).reset_index()

print(tabela_resumo)
```
