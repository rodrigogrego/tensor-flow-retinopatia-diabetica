```python
import array
import pandas as pd
import matplotlib.pyplot as plt
```

<h4>Grau de edema emacular (Treinamento)</h4>

```python
data = pd.read_csv(r'B.Disease Grading\2. Groundtruths\a.IDRiD_Disease Grading_Training Labels.csv')
data = data.drop(columns=['Image name', 'Unnamed: 3', 'Unnamed: 4', 'Unnamed: 5', 'Unnamed: 6', 'Unnamed: 7', 'Unnamed: 8', 'Unnamed: 9', 'Unnamed: 10', 'Unnamed: 11'])
data = data.groupby('Retinopathy grade').mean(numeric_only=True)
print(data)
plt.bar(data.index.tolist(), data['Risk of macular edema '])
```
