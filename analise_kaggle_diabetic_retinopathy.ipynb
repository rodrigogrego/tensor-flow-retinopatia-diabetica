```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
dgTest = pd.read_csv('test_labels.csv')
dgTrain = pd.read_csv('train_labels.csv')
```

```python
count_diagnosis_test = dgTest['diagnosis'].value_counts().sort_index()

cores = plt.cm.Reds(np.linspace(0.2, 1, 5)) 
count_diagnosis_test.plot(kind='bar', color=cores)

plt.title('ClassificaÃ§Ã£o das imagens analisadas labels_test')
plt.xlabel('ClassificaÃ§Ã£o')
plt.ylabel('Quantidade de imagens')

niveis_dr = ['NÃ£o possui RD', 'Leve', 'Moderada', 'Severa', 'RD Proliferativa']

plt.xticks(range(len(niveis_dr)), niveis_dr, rotation='horizontal')
```
